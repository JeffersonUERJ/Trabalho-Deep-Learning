## Entendendo os dados de soja(Importação e exportação)

### Bibliotecas



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Importando arquivo

```python
df_soja = pd.read_csv("C:/Trabalho Deep/src/data/Soja/V_EXPORTACAO_E IMPORTACAO_GERAL_2015-01_2025-12_DT20251115.csv",sep=';',encoding='utf8')
```

```python
df_soja.head(5)
```

```python
df_soja.columns
```

```python
#Excluindo a coluna Código NCM
df_soja.drop(columns={'Código NCM'},inplace=True)
```

```python
# Limpeza das colunas, retirando o caractere \r

colunas = ['Fluxo', 'Ano','Mês','Países','Descrição NCM','UF do Produto','URF','Valor US$ FOB','Quilograma Líquido']

for col in colunas:
    df_soja[col] = df_soja[col].astype(str).str.replace('\r','',regex=False)
```

```python
# Retirando caracteres vazios

for col in colunas:
    df_soja[col]= df_soja[col].str.strip()
```

```python
# Dividir os datasets em exportação e importação
df_soja_exp = df_soja[df_soja['Fluxo']=="Exportação"]
df_soja_exp
```

```python
df_soja_imp = df_soja[df_soja['Fluxo']=="Importação"]
df_soja_imp
```

```python
# Mantendo apenas o nome do mês na coluna mês
df_soja_exp['Mês'] = (
    df_soja_exp['Mês']
    .astype(str)
    .str.split('.')
    .str[1]
    .str.strip()
)
```

```python
df_soja_exp['URF'] = (
    df_soja_exp['URF']
    .astype(str)
    .str.split('-')
    .str[1]
    .str.strip()
)
```
