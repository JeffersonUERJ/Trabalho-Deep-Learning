### Bibliotecas

```python
import pandas as pd
import numpy as np
```

```python
df_exports = pd.read_excel('C:/Users/JeffersonKauanCavalc/OneDrive - ILOS/Documentos/Trabalho Deep/src/data/Soja/export.xls')
```

```python
estados_desejados = ['MT','RS','MT','PR','GO']
df_exports = df_exports[df_exports['UF/Regiões amparadas'].isin(estados_desejados)]
```

```python
df_exports['Vigência Final'] = (
    df_exports['Vigência Final']
    .astype(str)
    .str.split('-')
    .str[1]
    .str.strip()
)
```

```python
df_exports.rename(columns={'Vigência Final':'Ano'},inplace=True)
```

```python
df_exports.drop(columns=['Vigência Inicial'],inplace=True)
```

```python
df_exports = df_exports[df_exports['Ano']>='2020']
```
