### Desemprego

```python
# Desemprego
df_desemprego = pd.read_csv('C:/Users/JeffersonKauanCavalc/OneDrive - ILOS/Documentos/Trabalho Deep/src/data/Dados Externos/20251120033438.csv',sep=';',encoding='latin1',header=1)
```

```python
df_desemprego
```

```python
df_desemprego.columns
```

```python
df_longo = df_desemprego.melt(
    # Coluna que você quer manter fixa (a coluna 'Brasil' ou 'Local')
    id_vars=[' '], 
    
    # Nome da nova coluna que terá o nome do mês/ano (jan-2012, fev-2012, etc.)
    var_name='Ano_Mês', 
    
    # Nome da nova coluna que terá o valor da taxa de desemprego (8, 7.7, etc.)
    value_name='Taxa_Desemprego'
)

# Remova linhas que ainda contenham valores NaN na taxa de desemprego (dados que não existiam)
df_longo.dropna(subset=['Taxa_Desemprego'], inplace=True)
```

```python
df_longo.rename(columns={' ':'Local'},inplace= True
                )
```

```python
df_longo
```

```python
# 1. Extrair o Ano (Já feito anteriormente)
# Divide no espaço (último), pega a segunda parte
df_longo['Ano'] = df_longo['Ano_Mês'].str.split(' ').str[-1].astype(int)

# 2. Extrair a string do Mês (ex: 'jan-fev-mar')
df_longo['Meses_Rolling'] = df_longo['Ano_Mês'].str.split(' ').str[0]
```

```python
df_longo['Mês_Referencia_Nome'] = df_longo['Meses_Rolling'].str.split('-').str[0]

# 4. Converter o nome do mês para número para o merge
mapeamento_meses = {
    'jan': 1, 'fev': 2, 'mar': 3, 'abr': 4, 'mai': 5, 'jun': 6,
    'jul': 7, 'ago': 8, 'set': 9, 'out': 10, 'nov': 11, 'dez': 12
}
df_longo['Mês_numero'] = df_longo['Mês_Referencia_Nome'].map(mapeamento_meses)
```

```python
df_longo
```

```python
df_longo.drop(columns=['Ano_Mês','Meses_Rolling'],inplace=True)
```

```python
df_longo2 = df_longo[['Mês_numero','Mês_Referencia_Nome','Ano','Taxa_Desemprego']]
```

```python
df_longo2 = df_longo2[df_longo2['Ano']>=2020]
```

```python
df_desemprego2 = df_longo2.copy()
```

```python
df_desemprego2.tail()
```

```python
df_desemprego2.to_parquet('df_desemprego.parquet',index=False)
```

### DOLAR

```python
#Dolar
df_dolar = pd.read_csv('C:/Users/JeffersonKauanCavalc/OneDrive - ILOS/Documentos/Trabalho Deep/src/data/Dados Externos/STP-20251120115353227.csv',sep=';',encoding='latin1')
```

```python
df_dolar
```

```python
df_dolar['Data'] = pd.to_datetime(df_dolar['Data'], format='%d/%m/%Y',errors='coerce')
```

```python
# Cria uma nova coluna 'Mês' extraindo o número do mês
df_dolar['Mês'] = df_dolar['Data'].dt.month
```

```python
df_dolar['Ano'] = df_dolar['Data'].dt.year
```

```python
df_dolar['Nome_Mês'] = df_dolar['Data'].dt.strftime('%B')
```

```python
df_dolar.rename(columns={'1 - Taxa de câmbio - Livre - Dólar americano (venda) - diário - u.m.c./US$':'Valor'},inplace=True)
```

```python
df_dolar.dropna(subset=['Data'], inplace=True)
```

```python
df_dolar['Valor'] = df_dolar['Valor'].str.replace(',', '.', regex=False)
```

```python
df_dolar['Valor'] = df_dolar['Valor'].astype('float16')
```

```python
df_mensal = df_dolar.groupby(['Ano', 'Mês']).agg(
    # Calcula a média de uma coluna 'Cotação_Dolar'
    Media_Dolar=('Valor', 'mean'), 
).reset_index()
```

```python
df_mensal['Mês'] = df_mensal['Mês'].astype('int16')
```

```python
mês_numero = {
     1: 'Janeiro',
     2: 'Fevereiro',
     3: 'Março',
     4: 'Abril',
     5: 'Maio',
    6: 'Junho',
    7: 'Julho',
    8: 'Agosto',
    9: 'Setembro',
    10: 'Outubro',
    11: 'Novembro',
    12: 'Dezembro'
}
```

```python
df_mensal['Mês_Nome'] = df_mensal['Mês'].map(mês_numero)
```

```python
df_mensal
```

```python
df_mensal.to_parquet('df_mensal_dolar.parquet',index=False)
```
