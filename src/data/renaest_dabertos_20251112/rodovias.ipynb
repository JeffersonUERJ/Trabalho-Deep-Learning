## Frequência de acidentes

```python
# Supondo que você tenha a coluna 'Município'
df_acidentes_agrupado = df_periculosidade.groupby('uf_acidente').agg(
    Total_Acidentes=('qtde_acidente', 'sum'),
    Total_Obitos=('qtde_obitos', 'sum')
).reset_index()
```

```python
df_acidentes_agrupado
```

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np

# Cria uma cópia da coluna para evitar erro SettingWithCopyWarning
df_acidentes_agrupado['Acidentes_Norm'] = df_acidentes_agrupado['Total_Acidentes']

# Trata casos onde Acidentes é zero para evitar divisão por zero
df_acidentes_agrupado['Gravidade'] = np.where(df_acidentes_agrupado['Total_Acidentes'] == 0, 0, 
                           df_acidentes_agrupado['Total_Obitos'] / df_acidentes_agrupado['Total_Acidentes'])

# 1. Normaliza a Gravidade e a Frequência Bruta para a escala de 0 a 1
scaler = MinMaxScaler()
df_acidentes_agrupado['Frequencia_Norm'] = scaler.fit_transform(df_acidentes_agrupado[['Acidentes_Norm']])
df_acidentes_agrupado['Gravidade_Norm'] = scaler.fit_transform(df_acidentes_agrupado[['Gravidade']])

# 2. Combina as duas métricas com pesos iguais (50/50)
df_acidentes_agrupado['Índice_Risco_Simples'] = (0.50 * df_acidentes_agrupado['Frequencia_Norm']) + (0.50 * df_acidentes_agrupado['Gravidade_Norm'])
```

```python
df_acidentes_agrupado
```

```python
df_acidentes_agrupado['Classificacao_Risco'] = pd.qcut(
    df_acidentes_agrupado['Índice_Risco_Simples'],
    q=5, # Divide em 5 grupos (quintis)
    labels=[1, 2, 3, 4, 5], # Atribui os rótulos de 1 a 5
    duplicates='drop' # Ignora grupos se houver muitos valores idênticos
)
```

```python
df_acidentes_agrupado
```

```python
df_acidentes_simples = df_acidentes_agrupado.copy()
```

```python
df_acidentes_simples = df_acidentes_simples[['uf_acidente','Classificacao_Risco']]
```
